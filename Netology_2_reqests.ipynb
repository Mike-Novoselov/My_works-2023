```python
# Задача №1
# Кто самый умный супергерой?
import requests as rq

response = rq.get("https://cdn.jsdelivr.net/gh/akabab/superhero-api@0.3.0/api/all.json")
all_Heroes = response.json()
hero_names = ['Hulk', 'Captain America', 'Thanos']

# Поиск героев по списку имен и вывод их ума
for hero_name in hero_names:
    found_hero = False
    for hero in all_Heroes:
        if hero['name'] == hero_name:
            print("Ум героя", hero_name, "равен", hero['powerstats']['intelligence'])
            found_hero = True
            break
    if not found_hero:
        print("Герой с именем", hero_name, "не найден")

# Создание словаря для хранения умов героев
hero_intelligence = {}

# Добавление умов героев из списка в словарь
for hero in all_Heroes:
    hero_name = hero['name']
    if hero_name in hero_names:
        hero_int = hero['powerstats']['intelligence']
        hero_intelligence[hero_name] = hero_int

# Вывод словаря с умами героев
print(f"Самый умный {max(hero_intelligence)}")
```

```python
import requests

def upload_to_yandex_disk(file_path: str, disk_path: str):
    # Устанавливаем URL для загрузки файла
    upload_url = "https://cloud-api.yandex.net/v1/disk/resources/upload"

    # Устанавливаем параметры запроса
    params = {
        "path": disk_path, # Указываем путь на диске, куда нужно загрузить файл
        "overwrite": "true" # Указываем, что нужно перезаписать файл, если он уже существует
    }

    # Устанавливаем заголовки запроса
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"OAuth {MY_TOKKEN}" # Указываем токен авторизации
    }

    # Отправляем запрос на получение ссылки для загрузки файла
    response = requests.get(upload_url, headers=headers, params=params)

    # Если запрос успешен
    if response.status_code == 200:
        # Получаем ссылку для загрузки файла
        href = response.json()["href"]

        # Отправляем запрос на загрузку файла
        with open(file_path, "rb") as f:
            response = requests.put(href, files={"file": f})

            # Если запрос успешен
            if response.status_code == 201:
                print("Файл успешно загружен на Яндекс.Диск")
            else:
                print("Ошибка загрузки файла на Яндекс.Диск")
    else:
        print("Ошибка получения ссылки для загрузки файла на Яндекс.Диск")
```
